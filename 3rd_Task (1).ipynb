```python
#Imports
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle
import random
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
print("Import Done")
```

```python
#Loading the Data
dir='C:\\Users\\aiml1\\Desktop\\Prodigy\\CD'
categories=['CATS','DOGS']
data=[]
for category in categories:
    path=os.path.join(dir,category)
    label=categories.index(category)
    
    for img in os.listdir(path):
        imgpath=os.path.join(path,img)
        pet_img=cv2.imread(imgpath,0)
        try:
            pet_img=cv2.resize(pet_img,(50,50))
            image=np.array(pet_img).flatten()

            data.append([image,label])
        except Exception as e:
            pass
        
print(len(data))
```

```python
pick_in=open('data1.pickle','wb')
pickle.dump(data,pick_in)
pick_in.close() # ---> one run to save model
```

```python
pick_in=open('data1.pickle','rb')
data=pickle.load(pick_in)
pick_in.close()
```

```python
random.shuffle(data)
features=[]
labels=[]
for feature,label in data:
    features.append(feature)
    labels.append(label)
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,labels,test_size=0.25)
```

```python
#Model Evaluation
model=SVC(C=1,kernel='poly',gamma='auto')
model.fit(x_train,y_train)
pick=open('model1.sav','wb')
pickle.dump(model,pick)
pick.close()
```

```python
pick=open('model1.sav','rb')
model=pickle.load(pick)
pick.close()
```

```python
prediction=model.predict(x_test)
accuracy=model.score(x_test,y_test)
categories=['CATS','DOGS']
print('Accuracy: ',accuracy*100)
print('Prediction of randomly selected image from dataset: ',categories[prediction[0]])
mypet=x_test[0].reshape(50,50)
```
